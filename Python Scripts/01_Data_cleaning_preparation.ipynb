## Data Cleaning & Preparation:

    Dataset: Automobile_Sales_Analysis | 2020–2025
    Tool: Python (pandas, numpy)
    Goal: Load the raw dataset, assess data quality, fix issues, and export a clean file ready for analysis.

```python
import pandas as pd 
import numpy as np
```

### 1. Load the Dataset
We load the raw CSV and immediately check its shape and first/last rows. This confirms the file was read correctly and no rows were silently dropped.

```python
# LOAD DATA
#---------------------------------------------------
df = pd.read_csv("Automobile_Sales_Analysis.csv")

print("✓ File read successfully")

print("Shape :", df.shape)

print("First 3 rows:\n ", df.head(3))
print("Last 3 rows:\n", df.tail(3))

print("✓ Data loaded successfully!.")
#---------------------------------------------------
```

### 2. Explore Data Structure
Before cleaning anything, we understand the data — column types, non-null counts, and basic statistics to determine what needs fixing.

```python
# DATA EXPLORATION
#---------------------------------------------------
print("data information: \n")
print(df.info())

print("data summary: \n")
print(df.describe().round(2))

print("data types: \n")
print(df.dtypes)

for col in ['Region', 'Category', 'Fuel_Type', 'Dealer']:
    print(f"{col} ({df[col].nunique()} unique):")
    print(f"  {list(df[col].unique())}\n")
#---------------------------------------------------
```

### 3. Data Quality Checks
We check for four common data quality issues: **missing values**, **duplicate rows**, **data integrity** (Revenue = Units × Price), and **outliers**. Each issue is documented even if no fix is needed.

```python
# MISSING VALUES
#---------------------------------------------------
print("=== MISSING VALUES ===")
nulls = df.isnull().sum()

if nulls.sum() == 0:
    print("✓ No missing values found in any column.")
else:
    print("⚠ Missing values detected:")
    print(nulls[nulls > 0])
#---------------------------------------------------
```

```python
# DUPLICATE ROWS 
#---------------------------------------------------
print("=== DUPLICATE ROWS ===")
duplicates = df.duplicated().sum()
print(f"Duplicate rows found : {dupes}")

if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print(f"✓ Duplicates removed. New shape: {df.shape}")
else:
    print("✓ No duplicates. Dataset shape unchanged.")
#---------------------------------------------------
```

```python
# Data Integrity 
#---------------------------------------------------
# Revenue should equal Units_Sold × Unit_Price

print("=== REVENUE INTEGRITY CHECK ===")

df['Revenue_Expected'] = df['Units_Sold'] * df['Unit_Price']
df['Revenue_Match']    = np.isclose(df['Revenue'], df['Revenue_Expected'], rtol=0.01)

mismatch_count = (~df['Revenue_Match']).sum()
print(f"Rows where Revenue ≠ Units × Price : {mismatch_count}")

if mismatch_count == 0:
    print("✓ Revenue column is consistent. No integrity issues.")
else:
    print("Mismatches found — review below:")
    print(df[~df['Revenue_Match']][['Dealer','Vehicle_Model','Units_Sold','Unit_Price','Revenue','Revenue_Expected']])

df.drop(columns=['Revenue_Expected', 'Revenue_Match'], inplace=True)
print("✓ Temporary columns removed.")
#---------------------------------------------------
```

### 4. Fix Data Types & Extract Time Features
The **Date** column is stored as a string — we convert it to datetime, extract **Year**, **Month**, and **Quarter**.

```python
df['Date'] = pd.to_datetime(df['Date'])
print("✓ Successfully changed datatype of Date.")

print("For time_related columns: \n")
df['Year']    = df['Date'].dt.year
df['Month']   = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter
print("✓ Successfully created time-related columns!.")
```

### 5. Final Cleaned Dataset — Summary & Export
A final look at the clean dataset before saving.The cleaned file is exported as a new CSV — the original is never modified.

```python
df.to_csv("automobile_sales_cleaned.csv", index=False)

print("✓ Clean file saved → automobile_sales_cleaned.csv")
print("✓ Phase 1 complete. Ready for Phase 2 — SQL Analysis.")
```
